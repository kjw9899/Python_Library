## 1번 : 수치형 데이터 분포
### 문제
seaborn 을 이용해서 tips 데이터의 분포를 다음과 같이 4가지 방법으로 시각화 해보도록 합시다.

각각에 대하여 1, 2차원 히스토그램(histogram)과 밀도 도표(density plot)를 그리세요.

각 그래프의 이름은 [Plot 0, 1, 2, 3] =  [histplot, 2D histplot, kdeplot, 2D kdeplot] 입니다.

### 조건
* 색상별 데이터 밀도를 알 수 있게 colorbar를 붙히세요.
* matplotlib 기능을 이용해서 X, Y 축에 이름(label)을 붙이세요.
* 그림 크기(figsize)는 (6,5)로 지정하세요.
* 이름과 눈금(ticklabel)의 폰트 크기를 모두 크게(large) 합시다.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
# seaborn에 탑재된 tips 데이터셋을 사용합니다.

# Plot 0. "total_bill"
xlabel_0 = "Total Bill ($)"
ylabel_0 = "Count"
# X축에 "Total Bill ($)"이라는 레이블을, Y축에 "Count"를 달아줍니다.

# Plot 1. "tip"
xlabel_1 = "Tip ($)"
ylabel_1 = "Density"
# X축에 "Tip ($)"이라는 레이블을, Y축에 "Density"를 달아줍니다.

# Plot 2 & 3, "total_bill" vs "tip"
xlabel_2 = "Total Bill ($)"
ylabel_2 = "Tip ($)"
# X, Y축에 각기 "Total Bill ($)", "Tip ($)"이라는 레이블을 달아줍니다.

# label font
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# colormap
cmap = "Greens"
# 녹색 계열로 색칠합니다.

fig,ax = plt.subplots(figsize=(8,7),ncols=2, nrows=2)

sns.histplot(ax=ax[0,0], data = tips['total_bill'])
ax[0,0].set_xlabel(xlabel_0)
ax[0,0].set_ylabel(ylabel_0)

sns.kdeplot(ax=ax[0,1],data=tips['total_bill'])
ax[0,1].set_xlabel(xlabel_1)
ax[0,1].set_ylabel(ylabel_1)


sns.histplot(ax=ax[1,0], x=tips['total_bill'], y=tips['tip'], cbar=True, cmap=cmap)
ax[1,0].set_xlabel(xlabel_2)
ax[1,0].set_ylabel(ylabel_2)


sns.kdeplot(ax=ax[1,1],x=tips['total_bill'],y=tips['tip'],cmap=cmap)
ax[1,1].set_xlabel(xlabel_2)
ax[1,1].set_ylabel(ylabel_2)
```

## 2번 : 범주형 데이터 분포
### 문제
동일한 데이터를 이용해서, 수치 관점에서가 아닌, 범주의 관점에서 데이터의 분포를 그려봅시다.

다음과 같이 바 그래프(bar plot), 박스 플롯(box plot), 스웜플롯(swarm plot), 바이올린 플롯(violin plot)으로 그려보세요.

### 조건
* swarm plot과 violin plot은 남녀 성별에 따른 차이도 함께 그림에 포함되도록 하세요.

* 많은 그림을 그리기 위해서 3x2 축공간(axes)을 만들고, 그림 크기(figsize)는 10 x 12로 하세요.

* x 축을 공유해서 x 눈금 레이블(ticklabel)은 맨 아래에만 나오게 하고,

* x축과 y축 레이블은 각기 맨 아래와 왼쪽에 있는 그림들에만 붙입니다.

```python
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

tips = sns.load_dataset("tips")
# seaborn에 탑재된 tips 데이터셋을 사용합니다.

# label
xlabel = "Day"
ylabel = "Total Bill ($)"
# X축에 "Total Bill ($)"이라는 레이블을, Y축에 "Count"를 달아줍니다.

# label font
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# plot

fig, ax = plt.subplots(figsize=(10,12), nrows = 3, ncols= 2)

sns.barplot(ax = ax[0,0], x = "day", y = "total_bill", data = tips)
ax[0,0].set_ylabel(ylabel, label_fontdict, labelpad)
ax[0,0].set_xlabel("")

sns.boxplot(ax = ax[0,1], x = "day", y = "total_bill", data = tips)
ax[0,1].set_ylabel("")
ax[0,1].set_xlabel("")

sns.swarmplot(ax = ax[1,0], x = "day", y = "total_bill", data = tips)
ax[1,0].set_ylabel(ylabel, label_fontdict, labelpad)
ax[1,0].set_xlabel("")

sns.swarmplot(ax = ax[1,1], x = "day", y = "total_bill",hue="sex", dodge=True, data=tips)
ax[1,1].set_ylabel("")
ax[1,1].set_xlabel("")

sns.violinplot(ax = ax[2,0], x = "day", y = "total_bill", palette="muted", data=tips)
ax[2,0].set_ylabel(ylabel, label_fontdict, labelpad)
ax[2,0].set_xlabel(xlabel,label_fontdict, labelpad)

sns.violinplot(ax = ax[2,1], x = "day", y = "total_bill", data = tips)
ax[2,1].set_ylabel("")
ax[2,1].set_xlabel(xlabel,label_fontdict, labelpad)
```

## 3번 : 데이터 간 관계
### 문제
앞서 두 문제에서는 '분포'와 '범주'에 초점을 맞추었습니다. 이번에는 '관계' 에 초점을 두어봅시다.

다음 조건을 만족하는 산점도(scatter plot)와 선 그래프(line plot)를 그려보세요.

### 조건
* matplotlib 기능을 이용해서 X, Y 축에 이름(label)을 붙이세요.
* 이름과 눈금(ticklabel)의 폰트 크기를 모두 크게(large) 합시다.

```python
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn을 sns라는 이름으로 사용합니다.

tips = sns.load_dataset("tips")
# seaborn에 탑재된 tips 데이터셋을 사용합니다.

# label
xlabel = "Total Bill ($)"
ylabel = "Tip ($)"
label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}
labelpad = 12
# X, Y축에 각기 "Total Bill ($)"과 "Tip ($)" 라는 레이블을 달아줍니다.
# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.
# X, Y 축과의 거리는 12만큼 띄웁니다.

# ticks
tick_labelsize = "large"
# 눈금의 폰트 크기를 크게("large") 키웁니다.

# plot
fig, ax = plt.subplots(ncols=2, figsize=(10,4))

sns.scatterplot(ax = ax[0], x="total_bill", y="tip", data = tips)
ax[0].set_xlabel(xlabel, label_fontdict, labelpad)
ax[0].set_ylabel(ylabel, label_fontdict, labelpad)

sns.lineplot(ax = ax[1], x = "total_bill", y = "tip", data = tips)
ax[1].set_xlabel(xlabel, label_fontdict, labelpad)
ax[1].set_ylabel(ylabel, label_fontdict, labelpad)
```

## 4번 : 데이터 회귀선과 잔차
### 문제
이번에는 산점도(scatter plot)로 데이터의 추세(regression)를 알아보고, (regplot)

추세와 어긋나는 정도인 잔차(residual)를 그려봅시다. (residplot)

### 조건
* 추세선은 1차와 5차로 각각 그려보고, 5차는 회색 산점도에 오렌지색 추세선으로 그려봅니다.

* 신뢰구간(confidence interval)은 기본값인 95% 수준으로 표현합니다.

* 그림의 크기(figsize)는 (10,8)로 지정합니다.

```python
import matplotlib.pyplot as plt
import seaborn as sns


tips = sns.load_dataset("tips")

xlabel = "Total Bill ($)"
ylabel = "Tip ($)"
label_fontdict = {"fontsize": 14, "fontweight": "bold", "color": "gray"}
labelpad = 12

tick_labelsize = "large"


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axs = axes.ravel()

sns.regplot(data=tips, x='total_bill', y='tip', ax=axs[0], scatter=True, ci=95, order=1)
axs[0].set_xlabel(xlabel, label_fontdict, labelpad)
axs[0].set_ylabel(ylabel, label_fontdict, labelpad)

sns.regplot(data=tips, x='total_bill', y='tip', ax=axs[1],
            scatter=True, ci=95, order=5, scatter_kws={"color": "gray"}, line_kws={"color": "orange"})
axs[1].set_xlabel(xlabel, label_fontdict, labelpad)
axs[1].set_ylabel(ylabel, label_fontdict, labelpad)

sns.residplot(data=tips, x='total_bill', y='tip', ax=axs[2], order=1)
axs[2].set_xlabel(xlabel, label_fontdict, labelpad)
axs[2].set_ylabel(ylabel, label_fontdict, labelpad)

sns.residplot(data=tips, x='total_bill', y='tip', ax=axs[3], order=5, color='gray')
axs[3].set_xlabel(xlabel, label_fontdict, labelpad)
axs[3].set_ylabel(ylabel, label_fontdict, labelpad)
```

## 5번 : 데이터 간 상관도
### 문제
데이터셋을 이루고 있는 네 인자(feature) 간의 상관도를 볼 수 있도록,

페어플랏(pairplot) 기능을 사용하여 상관도를 그려보세요.

### 조건
* 가로 세로축에 4가지 인자들(Bill Length, Bill Depth, Flipper Length, Body Mass)을 나열하고,

  ① 대각선 (인자 이름이 일치하는)에는 데이터 분포를 쌓인(stack) 밀도 함수(density plot)으로 그립니다.

  ② 대각선 아래에는 산점도(scatter plot)와 추세선(regression)을 함께 표시합니다.

  ③ 대각선 위에는 2차원 밀도 함수(2D kdeplot)로 표현합니다.

* 그림 크기(figsize)는 (10,8)로 지정하세요.

* 이 때 눈금 레이블과 레이블의 폰트를 적절한 크기와 형식으로 조정해주세요.

```python
import matplotlib.pyplot as plt

# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.



import seaborn as sns

# seaborn을 sns라는 이름으로 사용합니다.



penguins = sns.load_dataset("penguins")

# seaborn에 탑재된 penguins 데이터셋을 사용합니다.

penguins["body_mass_100g"] = penguins["body_mass_g"]/100

# 펭귄 체중 데이터를 100g 단위로 표시합니다.

df = penguins.drop('body_mass_g',axis=1)



# label

labels=["Bill Length (mm)", "Bill Depth (mm)", "Flipper Length (mm)", "Body Mass (100g)"]

label_fontdict = {"fontsize":14, "fontweight":"bold", "color":"gray"}

labelpad = 12

# 폰트크기는 14, 두껍게("bold") 회색("gray")으로 씁니다.

# X, Y 축과의 거리는 12만큼 띄웁니다.



# ticks

tick_labelsize = "large"

# 눈금의 폰트 크기를 크게("large") 키웁니다.



# plot

g = sns.PairGrid(df, hue="species",diag_sharey=False)

g.map_diag(sns.kdeplot,fill=True)

g.map_lower(sns.scatterplot)

g.map_lower(sns.regplot, scatter=False, truncate=False, ci=False)

g.map_upper(sns.kdeplot, alpha=0.3)

g.add_legend()

for i in range(4):

    g.axes[3, i].set_xlabel(labels[i], fontdict=label_fontdict)

    g.axes[i, 0].set_ylabel(labels[i], fontdict=label_fontdict)

g.fig.align_ylabels(g.axes[:,0])

```
