## 3번 : 범위 지정과 scatter plot (객체지향)
### 문제
다음 조건을 만족하는 산점도(scatter plot)을 그리세요.

### 조건
* 제목(title)을 붙이세요.
* x축 범위(xlim)는 160~240 으로 지정하세요.
* y축 범위(ylim)를 2500~6500 으로 지정하세요.
* 객체지향 방식을 사용하세요.

### 데이터
seaborn 라이브러리에 내장되어있는 데이터인 penguin 체형 데이터를 사용합니다.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# seaborn 라이브러리를 sns라는 이름으로 사용합니다.

penguins = sns.load_dataset("penguins")
# seaborn에 내장된 펭귄 체형 데이터를 불러옵니다.

X = penguins["flipper_length_mm"]
# 펭귄 날개 길이를 X 변수로 지정합니다.

Y = penguins["body_mass_g"]
# 펭귄 체중을 Y 변수로 지정합니다.

title = "Flipper length(mm) vs Body mass(g)"
# 날개 길이와 체중의 상관관계를 그립니다.

fig, ax = plt.subplots()
ax.set_title(title)
ax.set_xlim(160,240)
ax.set_ylim(2500,6500)
ax.scatter(X,Y)
plt.show()
```

## 4번 : 크기 지정과 subplot (객체지향)

## 5번 : 꾸미고 이름달기, decorate, xlabel, ylabel
### 문제
이제부터는 객체지향 방식만을 사용해서, 그래프 하나의 완성도를 높여가보도록 합시다.

다음 조건을 만족하는 그래프를 완성하세요.

### 조건
* 다음과 같이 바 그래프(bar plot)를 그리고 X, Y 축에 이름(label)을 다세요.
* 바 그래프의 색상(color)과 선 색상(edgecolor)을 바꿔보세요.

## 6번 : 보조정보, text annotation
### 문제
다음과 같이 text annotation을 이용하여 보조적인 정보를 함께 보여줍시다. text를 붙여보세요.

### 조건
text는 검은색으로 굵게, Y값보다 4 낮은 위치에 center 정렬되어야 합니다.

```python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]
title = "Revenue (1000 $)"

fig, ax = plt.subplots(1,1)

ax.bar(X,Y, color = 'lightgray', edgecolor='black')
ax.set_xlabel('Salesperson', color = 'gray',fontsize=14, fontweight= 'bold')
ax.set_ylabel(title, color = 'gray', fontsize=14, fontweight= 'bold')

for i,v in enumerate(Y):
    ax.annotate(v, xy = (X[i],Y[i]), ha='center',va='bottom')
plt.show()
```

## 7번 : 보조정보, axline, shade
### 문제
이번에는 수평선과 구간 음영이라는 보조정보를 그림에 추가해봅시다. line과 음영을 삽입하세요

```python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]
title = "Revenue (1000 $)"

fig, ax = plt.subplots(1,1)

ax.bar(X,Y, color = 'lightgray', edgecolor='black')
ax.set_xlabel('Salesperson', color = 'gray',fontsize=14, fontweight= 'bold')
ax.set_ylabel(title, color = 'gray', fontsize=14, fontweight= 'bold')

for i,v in enumerate(Y):
    ax.annotate(v, xy = (X[i],Y[i]), ha='center',va='bottom')
    
ax.axhline(45, color ='b', linestyle= '--', linewidth = 2)
ax.axhline(15, color ='r', linestyle= '--', linewidth = 2)
ax.axhspan(0,15, facecolor='red', alpha=0.2)

plt.show()
```

## 8번 : 데이터 잘 보여주기, grid, spine
### 문제
데이터를 잘 보여줄 수 있도록 하기 위해서 그래프의 배경에 해당하는 요소들을 다듬어봅시다.

테두리(spines)를 아래만 남기고 지우고, 아래 테두리는 두껍게 한 후 배경에 눈금(grid)을 붙이세요.

```python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]
title = "Revenue (1000 $)"

fig, ax = plt.subplots(1,1)

ax.bar(X,Y, color = 'lightgray', edgecolor='black')
ax.set_xlabel('Salesperson', color = 'gray',fontsize=14, fontweight= 'bold')
ax.set_ylabel(title, color = 'gray', fontsize=14, fontweight= 'bold')

for i,v in enumerate(Y):
    ax.annotate(v, xy = (X[i],Y[i]), ha='center',va='bottom')
    
ax.axhline(45, color ='b', linestyle= '--', linewidth = 2)
ax.axhline(15, color ='r', linestyle= '--', linewidth = 2)
ax.axhspan(0,15, facecolor='red', alpha=0.2)

plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거
ax.spines['bottom'].set_lw(2)

plt.grid(True, axis='y') # y축 격자만 생성


plt.show()
```

## 9번 : 눈금 간격 조정하기, tick
### 문제
뒤의 배경에 넣은 grid 가 너무 빼곡하게 들어가있는 느낌이 들지는 않나요?

한번 grid 간격을 수정해봅시다.

### 조건
y축 눈금(tick)을 15 간격으로 조정하고 보이지 않도록(length=0) 하세요.

눈금의 폰트 크기도 크게(large) 하세요.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]
title = "Revenue (1000 $)"

fig, ax = plt.subplots(1,1)

ax.bar(X,Y, color = 'lightgray', edgecolor='black')
ax.set_xlabel('Salesperson', color = 'gray',fontsize=14, fontweight= 'bold')
ax.set_ylabel(title, color = 'gray', fontsize=14, fontweight= 'bold')

for i,v in enumerate(Y):
    ax.annotate(v, xy = (X[i],Y[i]), ha='center',va='bottom')
    
ax.axhline(45, color ='b', linestyle= '--', linewidth = 1)
ax.axhline(15, color ='r', linestyle= '--', linewidth = 1)
ax.axhspan(0,15, facecolor='red', alpha=0.2)

plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거
ax.spines['bottom'].set_lw(2)

plt.grid(True, axis='y') # y축 격자만 생성

ax.yaxis.set_major_locator(ticker.MultipleLocator(15)) # 축 그리드 간격 조절
ax.tick_params(axis = 'x', length = 0)
ax.tick_params(axis = 'y', length = 0, labelsize = 'large')

plt.bar(X[4], Y[4], color = ['yellow'], edgecolor='red', alpha = 1)



plt.show()
```
