## 1번 : 기본적인 line plot

주어진 베이스라인을 이용해 다음과 같은 선 그래프(line plot)을 그리세요.

## 2번 : 제목과 bar plot

주어진 베이스라인을 이용해 다음과 같은 바 그래프(bar plot)을 그리고 제목(title)을 붙이세요.

```python
#data
X = [0, 1, 2, 3, 4]
Y = [10, 20, 50, 40, 60]
title = "Revenue (1000 $)"

plt.bar(X,Y)
plt.title(title)
plt.xlabel('X',fontsize=15)
plt.ylabel('Y',fontsize=15)
plt.show()
```

## 3번 : 범위 지정과 scatter plot

다음 조건을 만족하는 산점도(scatter plot)을 그리세요.

```python
import matplotlib.pyplot as plt
# matplotlib의 pyplot모듈을 plt라는 이름으로 사용합니다.

import seaborn as sns
# seaborn 라이브러리를 sns라는 이름으로 사용합니다.

penguins = sns.load_dataset("penguins")
# seaborn에 내장된 펭귄 체형 데이터를 불러옵니다.

X = penguins["flipper_length_mm"]
# 펭귄 날개 길이를 X 변수로 지정합니다.

Y = penguins["body_mass_g"]
# 펭귄 체중을 Y 변수로 지정합니다.

title = "Flipper length(mm) vs Body mass(g)"
# 날개 길이와 체중의 상관관계를 그립니다.

plt.scatter(X,Y)
plt.title(title)
plt.xlabel('flipper_length')
plt.ylabel('body_mass')
plt.xlim(160,240)
plt.ylim(2500,6500)
plt.show()
```

## 4번 : 크기 지정과 subplot

### 문제
주어진 동일한 데이터를 이용하여 다음과 같은 3가지 그래프들을 그리세요.

### 조건
* 그림 크기(figsize)를 9 x 3으로 지정하고, 가로로 세 개의 그래프를 한 줄로 배치하세요.
* 차례대로 산점도(scatter plot), 바 그래프(bar plot), 선 그래프(line plot)로 표현하세요.
* pyplot 방식을 사용하세요.

```python
X = [0, 1, 2, 3, 4]
Y = [10, 20, 50, 40, 60]

plt.figure(figsize=(9,3))
plt.subplot(1,3,1)
plt.scatter(X, Y)

plt.subplot(1,3,2)
plt.bar(X,Y)

plt.subplot(1,3,3)
plt.plot(X,Y)
```
