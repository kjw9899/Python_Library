## 23.aggregating
각 row(행)의 평균을 저장하고 있는 column(열)을 만들고, column의 이름을 "평균"으로 지정하는 코드를 작성하세요.

```python
df = {'A' : range(1,6),
      'B' : [-1,-2,-3,-4,-5],
      'C' : [2,3,4,7,8]
      }
df = pd.DataFrame(df)

df['평균'] = df.mean(axis=1)
df
```

## 25.aggregating (3)

생성된 DataFrame에서 label A와 label B를 기준으로 val1, val2, val3의 합계를 구하는 코드를 작성해보세요.

```python
df_25 = {'A' : range(1,6),
         'B' : reversed(range(-5,0)),
         'C' : [2,3,4,7,8],
         'label' : ['A','A','B','B','B']}

df_25 = pd.DataFrame(df_25)
df_25

df_25 = df_25.groupby(["label"]).sum()
df_25
```

## 26.pivot table

직전 문제와 동일한 DataFrame 을 이용하여, label column을 기준으로 pivot table(피벗테이블)을 생성하는 코드를 작성해보세요.

```python
df_25 = df_25.pivot_table(index = 'label')
df_25
                       
```
